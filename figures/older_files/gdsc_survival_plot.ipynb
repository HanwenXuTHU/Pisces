```python
import os
import sys
import fig_settings, fig_utils
import matplotlib.pyplot as plt
from tqdm import tqdm

root_dir = '../Pisces/scripts/case_study/gdsc_novel_combo/'
sys.path.append(root_dir)
from utils import search_patients_typing_drug_combo, load_survival_info

ct='COADREAD'
fig_path = f'save_figs/typing_survival_plot/{ct}/'
if not os.path.exists(fig_path):
    os.makedirs(fig_path)

name2group, name2edges = search_patients_typing_drug_combo(ct=ct, th=30)
for n in tqdm(name2group.keys()):
    case, control = name2group[n][0], name2group[n][1]
    if len(case) == 0:
        continue
    res = load_survival_info(case, control, ct=ct)
    ax = fig_settings.get_wider_axis()
    fig_utils.kaplan_meier_curve(ax, res['survival time'], res['survival status'],
                                ['Case', 'Control'], xlabel='Time (months)',
                                ylabel='Survival probability', showCI=False)
    plt.title(ct)
    fig_utils.format_ax(ax)
    plt.savefig(f'{fig_path}/{ct}_{n}_survival_plot.png', bbox_inches='tight')
```

```python
import os
import sys
import fig_settings, fig_utils
import matplotlib.pyplot as plt
from tqdm import tqdm

root_dir = '../Pisces/scripts/case_study/gdsc_novel_combo/'
sys.path.append(root_dir)
from utils import search_patients_typing_drug_combo, load_survival_info

ct='BRCA'
n = 'BT-20_PF-4708671_Afatinib'
fig_path = f'save_figs/typing_survival_plot/{ct}/'
if not os.path.exists(fig_path):
    os.makedirs(fig_path)

name2group, name2edges = search_patients_typing_drug_combo(ct=ct, th=30)

case, control = name2group[n][0], name2group[n][1]
res = load_survival_info(case, control, ct=ct)
ax = fig_settings.get_wider_axis()
fig_utils.kaplan_meier_curve(ax, res['survival time'], res['survival status'],
                            ['Case', 'Control'], xlabel='Time (months)',
                            ylabel='Survival probability', showCI=False)
plt.title(ct)
fig_utils.format_ax(ax)
plt.savefig(f'{fig_path}/{ct}_{n}_survival_plot.png', bbox_inches='tight')
# save the pathway
all_edges = name2edges[n]
save_path = f'{root_dir}/output/{ct}_{n}_pathway.txt'
f = open(save_path, 'w')
for e in all_edges:
    f.write(f'{e[0]}-->>{e[1]}\n')
```
